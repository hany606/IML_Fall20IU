```python
# import the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import Lasso, Ridge

%matplotlib inline
```

```python
# Synthesize the training data and the test data
rng = np.random.RandomState(1)
x = 10*rng.rand(100)
y = np.sin(x) + 0.1 * rng.randn(100)
X_test = np.linspace(0, 10, 100)
```

```python
# Plot the training data for visualization
plt.scatter(x, y, marker=".", label="Train-data")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Task1-Data")
plt.show()
```

### Polynomial Regression Model without Regularization

The best model is the one that has the least MSE with the least standard deviation of the cross-validation folds

```python
degrees = [1, 4, 6, 7, 8, 9, 12, 13, 15] # choose some degrees that one of them maybe the best model
plt.figure(figsize=(20, 15))


# Function for creation of the pipeline for the regression with fitting to the training data of this task
def polynomial_regression_pipeline(deg):
    polynomial_features = PolynomialFeatures(degree=deg, include_bias=False)
    linear_regression = LinearRegression()
    pipeline = Pipeline([("polynomial_features", polynomial_features),
                         ("linear_regression", linear_regression)])
    pipeline.fit(x[:, np.newaxis], y) # fit the pipeline to the training data
    # Evaluate the models using crossvalidation
    scores = cross_val_score(pipeline, x[:, np.newaxis], y, scoring="neg_mean_squared_error", cv=10)
    return pipeline, scores

# Lists to store the MSE and STD for different degrees of the model
mse = []
std = []
# Iterate over the list of the chosen degrees
for i, deg in enumerate(degrees):
    # Related to the shape of the plotting figure
    rows = len(degrees)//3   # number of rows
    cols = len(degrees)//rows   # number of cols
    ax = plt.subplot(rows, cols, i+1)   # plot in the ith part from the figure
    plt.setp(ax, xticks=(), yticks=())  # Remove the line number of the figures as we care about only the visualization of the model not the exact number of the number line
    
    pipeline, scores = polynomial_regression_pipeline(deg)   # create, fit, and get the scores of the pipeline
    
    y_test_pred = pipeline.predict(X_test[:, np.newaxis])   # predict the new y from the test data
    
    # Plot the test data with its predictions, and the training data
    plt.plot(X_test, y_test_pred, color='orange', label= "Model")
    plt.scatter(x, y, marker='.', label="Samples")
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.legend(loc="best")
    mse.append(-scores.mean())
    std.append(scores.std())
    plt.title("Degree {}\nMSE = {:.2e}(+/- {:.2e})".format(deg, -scores.mean(), scores.std()))
plt.show()
```

The model with 8 degrees appeared to have the least MSE, however, the model with 6 degrees has a little bigger MSE than the model of 8 degrees but has less STD.

I will choose the model with 8 degrees as the best model, however, I will test both of them with regularization and see how much the model can be improved with regularization.

```python
# printing the MSE and STD of each iteration over the different degrees
for i,deg in enumerate(degrees):
    print("{:}. Model with degree({:}) -> MSE = {:.2e}, std = {:.2e}".format(i+1, deg, mse[i], std[i]))
```

```python
# Plot the MSE and STD vs degrees
plt.xticks(degrees)
plt.plot(degrees, mse, label= "MSE(deg)")
plt.plot(degrees, std, label= "std(deg)")
plt.xlabel('Degree')
plt.ylabel('MSE/std')
plt.legend(loc="best")
plt.show()
```

### Polynomial Regression Model with Regularization

```python
degrees = [6, 8]    # chosen degrees as the best models to use regularization with 
plt.figure(figsize=(20, 10))

# Function for creation of the pipeline for the regression with regularization with fitting to the training data of this task
def regularization_pipeline(regularizor, degree):
    polynomial_features = PolynomialFeatures(degree=degree, include_bias=False)
    pipeline = Pipeline([("polynomial_features", polynomial_features),
                         ("regularizor", regularizor)])
    pipeline.fit(x[:, np.newaxis], y) # fit the pipeline to the training data
    # Evaluate the models using crossvalidation
    scores = cross_val_score(pipeline, x[:, np.newaxis], y, scoring="neg_mean_squared_error", cv=10)
    return pipeline, scores

# Iterate over the list of the chosen degrees
for i, deg in enumerate(degrees):
    rows = 1
    cols = 2
    ax = plt.subplot(rows, cols, i+1)
    plt.setp(ax, xticks=(), yticks=())
    
    regression, scores = polynomial_regression_pipeline(deg)
    lasso, lasso_scores = regularization_pipeline(Lasso(alpha=0.001), deg)
    ridge, ridge_scores = regularization_pipeline(Ridge(alpha=0.1), deg)

    y_test_pred = regression.predict(X_test[:, np.newaxis])
    y_test_pred_lasso = lasso.predict(X_test[:, np.newaxis])
    y_test_pred_ridge = ridge.predict(X_test[:, np.newaxis])

    print("{:}. Model with degree({:}) -> MSE = {:.2e}, std = {:.2e}".format(1, deg, -scores.mean(), scores.std()))
    print("{:}. Model with degree({:}), with Lasso -> MSE = {:.2e}, std = {:.2e}".format(2, deg, -lasso_scores.mean(), lasso_scores.std()))
    print("{:}. Model with degree({:}), with Ridge -> MSE = {:.2e}, std = {:.2e}".format(3, deg, -ridge_scores.mean(), ridge_scores.std()))
    print("---------------------------")
    
    plt.plot(X_test, y_test_pred, color='green', label= "Model-no-regularization")
    plt.plot(X_test, y_test_pred_lasso, color='orange', label= "Model (Lasso)")
    plt.plot(X_test, y_test_pred_ridge, color='red', label= "Model (Ridge)")
    plt.scatter(x, y, marker='.', label="Samples")
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.xticks()
    plt.legend(loc="best")
    plt.title("Degree {}\nMSE = {:.2e}(+/- {:.2e})\nLasso_MSE = {:.2e}(+/- {:.2e})\nRidge_MSE = {:.2e}(+/- {:.2e})".format(deg, -scores.mean(), scores.std(), -lasso_scores.mean(), lasso_scores.std(),-ridge_scores.mean(), ridge_scores.std()))

plt.show()
```

For the model that has degree 6: Lasso didn't improve the model from the point of view of MSE, and Ridge improved slightly the model.

For the model that has degree 8: Lasso didn't improve the model from the point of view of MSE, but Ridge improved the model significantly the model.

The best model is the one that has degree 8 as polynomial regression with Ridge (L2) regularization.

### Preprocessing the data

#### Step 1: Import the libraries

```python
# import the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import RobustScaler
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score

%matplotlib inline
```

#### Step 2: Import the dataset

```python
# Reading and loading the data from the csv file inside data directory
data = pd.read_csv('./data/banking.csv')
```

```python
# Display the data to know what is this data about; For more information: http://archive.ics.uci.edu/ml/datasets/Bank+Marketing
data.loc[:,:"nr_employed"].head(5)
```

```python
# [Not necessary step as the best model exceeds 0.91 test accuracy without this step]
# Drop unncessary data: contact, month, day_of_week, (default, : not sure)
# new_data = data.drop(["contact", "month", "day_of_week"], axis=1)
new_data = data
```

```python
X = new_data.loc[:, :'nr_employed']   # take the columns except y column
Y = new_data['y']    # take only the y column for the label
```

```python
# Display part of the data for visualization and understanding
# data.head(10)
X.head(5)
# Y.head(5)
```

#### Step 3: Encode the Categorical data

Label Encoding could be used in the case of education as university degree is better than high school (ordinal categorical features) (Not necessary step as the model has exceeded 0.9 test accuracy)

And one-hot encoding is being used for the nominal categorical features

Good source for label encoding vs one-hot encoding: https://towardsdatascience.com/categorical-encoding-using-label-encoding-and-one-hot-encoder-911ef77fb5bd

```python
# One-hot encoding for poutcome, (day_of_week, month, contact (maybe deleted)), marital, job, default, housing, loan
ohe_encoder = OneHotEncoder(handle_unknown="ignore", sparse=False)

ohe_features = ["poutcome", "education", "marital", "job", "default", "housing", "loan", "day_of_week", "month", "contact"]

ohe_encoder.fit(X[ohe_features])

def ohe_encoding(df, features_name, encoder):
    new_feats = encoder.transform(df[features_name])
    new_cols = pd.DataFrame(new_feats, dtype=int, columns=encoder.get_feature_names(features_name))
    new_df = pd.concat([df, new_cols], axis=1)    
    new_df.drop(features_name, axis=1, inplace=True)
    return new_df

X = ohe_encoding(X, ohe_features, ohe_encoder)
```

```python
X.head(5)
```

#### Step 4: Check out the missing values and impute them using the mean value strategy.

```python
# Imputation
def count_nans(df):
    return pd.isna(df).sum().sum()

print("Number of nans in X: {:}".format(count_nans(X)))
imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
imp_mean.fit(X)
X_imputed = imp_mean.transform(X)
print("Number of nans in X: {:}".format(count_nans(X_imputed)))
```

#### Step 5: Feature scaling

```python
# Feature scalling: using robust scaler, it uses similar technique as minmax but with quartile kind of filter to remove the outliers
# Good source for Feature scaling: https://benalexkeen.com/feature-scaling-with-scikit-learn/
robust_scalar = RobustScaler()
robust_scalar.fit(X_imputed)
X_scaled = robust_scalar.transform(X_imputed)
# scaler = MinMaxScaler()
# scaler.fit(X_imputed)
# X_scaled = scaler.transform(X_imputed)
```

#### Step 6: Splitting the dataset into Training (80 %) and Test set (20 %)

```python
# Splitting the data with fixed random seed to reproduce the testing and trainind datasets with stratify on the labels to have balanceed datasets 
x_train, x_test, y_train, y_test = train_test_split(X_scaled, Y, test_size=0.2, stratify=Y, random_state=123)
```

```python
print("% of test data to all data:", len(x_test)/(len(x_train)+len(x_test)))
print('% of positive samples in train set:', sum(y_train == 1) / len(y_train))
print('% of positive samples in test set:', sum(y_test == 1) / len(y_test))
```

### Build the model

```python
# Set the possible values for the parameters size of the penalty(l1,l2), type of solver: (liblinear, lbfgs), classifier_C: np.logspace(−4, 4, 20)
param_grid = {'penalty': ["l1", "l2"],
              'solver': ["liblinear", "lbfgs"],
              'C': np.logspace(-4, 4, 20)}

# Use logistic regression with gridsearchcv that has cross-validation with grid search over the hyper-parameters of the algorithm
grid_search_clf = GridSearchCV(LogisticRegression(), cv=10
                               ,param_grid=param_grid, scoring='accuracy')

grid_search_clf.fit(x_train, y_train)   # fit the gridsearch on the training data

# get the mean and std values of the test scores for the accuracy
means = grid_search_clf.cv_results_['mean_test_score']
stds = grid_search_clf.cv_results_['std_test_score']    

# print all the combinations that gridsearch has been running on it
for mean, std, params in zip(means, stds, grid_search_clf.cv_results_['params']):
    print("{:.3f} (+/-{:.3f}) for {:}".format(mean, std * 2, params))

# print the best parameters to be used later
print("\nBest parameters set found on development set:")
print(grid_search_clf.best_params_)
```

```python
# create the logistic regression model with the best parameters that has been found during the gridsearch
# clf_best = LogisticRegression(C=1.623776739188721, penalty="l2", solver="liblinear")
# clf_best.fit(x_train, y_train) # fit it with the training data
# y_pred = clf_best.predict(x_test)

# using the best model from the grid search on the logistic regression, to predict the test data and check its:
# accuracy, precision, recall scores
y_pred = grid_search_clf.predict(x_test)
print("Accuracy: {:.5f}".format(accuracy_score(y_test, y_pred)))
print("Precision: {:.5f}".format(precision_score(y_test, y_pred)))
print("Recall: {:.5f}".format(recall_score(y_test, y_pred)))
```
