### 1.1 Simple NN Search

Implement the simplest form of NN search (exhaustive search).
Your function should: 
1. Calculate the distance between the test point and each dataset point using euclidean distance.
2. Sort the dataset according to calculated distance in ascending order.
3. Return the top K points from the sorted dataset.

You need to implement this from scratch. Don't use library solutions. 
It is okay to use library function for eucliden distance.

```python
def simple_nns(v, data, k):
    """
    param: v: test point
    param: data: dataset
    param: k: number of nearest neighbors to retrive
    return: list of nearest points from the dataset
    """
    # implement simple NN Search

    # 1. Calculate the distance between the test point "v" and each point in the dataset "data"
    distances_points = [] # It will store the distances between the points and the indices of the point from the dataset
    for i,d in enumerate(data):
      distance = np.linalg.norm(np.array(v)-np.array(d))  # Calculate the l2 norm = euclidain distance between the test point and a point in the dataset
      distances_points.append({"distance": distance, "index":i})
    # 2. Sort the distances list in ascending order throught the built-in function in python lists
    distances_points.sort(key= lambda x: x["distance"])
    # 3. Get the K (top) nearest points
    k_top_distances = distances_points[:k]
    return np.array(k_top_distances)
```

### 1.2 Improving NNS

Choose any method you like to speed up NN search, understand it thoroughly and learn how to apply it. Explain the way it works in few sentences and in your own words.

As it was described in the assignment's description, there are two categories for the Nearest Neighbour Search: The Exact and the Approximate methods. The approximate can be used efficiently with big datasets better than the exact methods as it will provide approximate results but not the exact but in better performance (trade-off between the performance and the quality of the results).

Here, I have decided to try one from the exact methdos: KD-Tree and one from the approximate methods: ANNOY from Spotify and compare the performance of each one of them and check which one will perform better according to certain Hyper-parameters that can be tuned later.


#### KD-tree:
* It is one of the data strucutres that efficiently represent the data through partioning based on specific conditions. It is organizing the points(dataset) in k-dimensional space. It is represented as binary tree with k-dimensional point in every leaf.

* Building the tree: O(D*NlogN) such that D is the dimension of the single data in terms of sklearn explanation, but building the tree is done once on the whole dataset at the beginning.

    - Adding node by node according to the condints of the branch to be added right or left of the root node of the sub-tree

* Query the tree: O(NlogN)
    - Traverse the tree starting from the root till the point that is requested to find its neighbours, Traverse up and down to find the neighbours according to specific algorithm in Source2

#### ANNOY:
* It is a library from Spotify, it is Approximate Nearest Neighbour optimized for memory usage. According to the creators on their [Github](https://github.com/spotify/annoy) "Annoy is almost as fast as the fastest libraries".

* It is using static files as indexes and it can share index across processes which can enhance the performance. It builds the index once and can be used later.

* It also creates large read-only file-based data structures that can be shared by many processes

* It is based on building forest of trees. More trees gives higher precision when querying

* It should be faster within the big dataset as it doesn't provide exact but an approximate result.

## 2. Gaussian Naive Bayes (4 points)

Finish the implementation of Gaussian NB.

```python
# Sources: https://iq.opengenus.org/gaussian-naive-bayes/ , https://www.saedsayad.com/naive_bayesian.htm#:~:text=Naive%20Bayes%20classifier%20assume%20that,the%20values%20of%20other%20predictors.&text=P(c%7Cx)%20is,the%20prior%20probability%20of%20class.
class GaussianNB:
    def fit(self, X_train, y_train):
        """
        fit classifier, i.e. calculate needed statistics
        """
        ### START CODE HERE ###
        self.prior = [] # P(Class)
        self.means = [] # Mean of each feature per class 
        self.std = []   # STD of each feature per class
        self.unique_classes = np.unique(y_train)  # set of unique classes
        for c in self.unique_classes:
          self.prior.append(sum([(1 if c == v else 0)for v in y_train])/len(y_train)) # calculate the prior probability: sum(class(i)==c)/num_data
          data_class = [] # data per class: collect the features of the entry(row) that belongs to the c class that is being iterated over
          # iterate over the features of the dataset
          for j,x in enumerate(X_train):
            if(c == y_train[j]): # if the jth row belongs to the c class, then append it
              data_class.append(x)
          self.means.append(np.mean(data_class,axis=0)) # calculate the mean for every feature per c class
          self.std.append(np.std(np.array(data_class),axis=0)) # calculate the std for every feature per c class
        # print(np.array(self.means).shape)
        # print(self.std)
        ### END CODE HERE ###
        
    def predict(self, X_test):
        """
        predict class labels for test data points
        param: test_data: list of data points to be classified
        return: predictions (class labels)
        """
        ### START CODE HERE ###
        predictions_proba_data = [] # collect the probability for each entry (data/row) in the testing dataset
        predictions = []  # collect the predicted class for each entry (data/row) in the testing dataset
        # Iterate over all the testing dataset
        for x in X_test:
          predictions_proba_data.append([])
          # Iterate over each class and calculate its probability
          for i in range(len(self.unique_classes)):
            prediction_proba_class = 1 # it will store the likelihood probability of the data for the ith class
            # Iterate over the features inside the data to calculate the likelihood probability P(x|c_i) = P(x_0|c_i)*P(x_1|c_i)*...*P(x_n|c_i)
            for j,x_feature in enumerate(x):
              # Apply gaussian distribution to find the likelihood probability P(x_j|c_i)
              prediction_likelihood = (np.exp((-(x_feature-self.means[i][j])**2)/(2*(self.std[i][j]**2)))/(np.sqrt(2*np.pi*(self.std[i][j]**2))))
              prediction_proba_class *= prediction_likelihood
            # Calculate the probability of the class = P(x|c_i)*P(c_i)
            prediction_proba_class *= self.prior[i]
            predictions_proba_data[-1].append(prediction_proba_class) # append the probability of the entry to belong to the ith class to the corresponding entry in the testing dataset
          # Get the argument that has the maximum probability and this is the class that this entry belongs to
          predictions.append(np.argmax(predictions_proba_data[-1]))
        ### END CODE HERE ###
        return predictions
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.datasets import load_iris
# Improt the library for the original GaussianNB to use it as Benchmark and renaming it to something else
from sklearn.naive_bayes import GaussianNB as GBS

iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.5, random_state=0)  # split the dataset

# Test my implementation
gnb = GaussianNB()
gnb.fit(X_train, y_train)
y_pred = gnb.predict(X_test)

# Test the orignal implementation of sklearn
gnb_sci = GBS()
gnb_sci.fit(X_train, y_train)
y_pred_scklearn = gnb_sci.predict(X_test)

# You should get the accuracy score not less than 90%.
print(accuracy_score(y_test, y_pred))

print(f"Accuracy of the original Scikit-learn implementation: {accuracy_score(y_test, y_pred_scklearn):.3f}, Accuracy of my implementation {accuracy_score(y_test, y_pred):.3f}")
```

## 3. PCA

Apply PCA to the Digits dataset, produce the scree plot for the given data and decide the optimal number of PCs to keep. Calculate and print their total explained variance.

```python
from sklearn.datasets import load_digits
# Import PCA library from sklearn as in the lab
from sklearn.decomposition import PCA

X, y = load_digits(return_X_y=True)

pcs = [i for i in range(1,21)]  # make a list of needed number of principle components to be used as x-axis for the plot later

# apply PCA
pca = PCA(n_components=pcs[-1]) # Apply PCA for the needed number of components as it will return of vectors for each number of component
x_PCA = pca.fit_transform(X)
pcs_explained_variance = pca.explained_variance_ratio_*100

# make scree plot
fig = plt.figure(figsize=(8,5))
plt.plot(pcs, pcs_explained_variance, 'o-', c="black", linewidth=2)
plt.title('Scree Plot for Principle Components')
plt.xlabel('Principal Components')
plt.ylabel('Proportion of Variance Explained (%)')
plt.xticks(pcs)
plt.grid()
plt.show()

# choose the number PCs and calculate and print their total explained variance
chosen_num_components = 11
pca = PCA(n_components=chosen_num_components)
x_PCA = pca.fit_transform(X)
print(f"Total Explained Variance for (num_componenets = {chosen_num_components}): {pca.explained_variance_.sum()}")
```
